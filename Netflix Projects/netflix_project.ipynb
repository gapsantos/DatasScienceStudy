```python
# Importing pandas and matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")


print(netflix_df['duration'])
```

## First Question


What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

# 

```python
# Start coding here! Use as many cells as you like

decade90s = np.logical_and(netflix_df['release_year'] >= 1990, netflix_df['release_year'] < 2000)

movies_1990 = netflix_df[decade90s]


# to understand the data :print(movies_1990['duration'])

#ploting the data in an histogram

#?plt.hist

plt.hist(movies_1990['duration'], bins = 30)

plt.show()

#What was the most frequent movie duration in the 1990s?

#?np.unique

vals, counts = np.unique(movies_1990['duration'], return_counts = True)

#print(counts)

#?np.argwhere

mode_value = vals[np.argwhere(counts == np.max(counts)).flatten()]

duration = int(mode_value)

print(duration)

print(movies_1990)
```

## Second Question

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# Count the number of short action movies released in the 1990s and save this integer as short_movie_count

short = movies_1990['duration'] < 90
action = movies_1990['genre'] == 'Action'

short_action = np.logical_and(short,action)

print(short_action)

short_action_movie = movies_1990[short_action]

print(short_action_movie)


short_movie_count = 0


for label,row in short_action_movie.iterrows():
    short_movie_count = short_movie_count + 1

print(short_movie_count)
```
